```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(bars_in_cities))
```

```python
# Create a bar chart based upon the above data
plt.bar(cities, bars_in_cities, color='b', alpha=0.5, align="center")
```

```python
# Create the ticks for our bar chart's x axis
# Action for Variable in Container
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"] )
```

```python
# Set the limits of the x axis
plt.xlim(-0.75, len(x_axis)-0.25)
```

```python
# Set the limits of the y axis
plt.ylim(0, max(bars_in_cities)+1)
```

```python
# Give the chart a title, x label, and y label

plt.title("Bars in US Cities")
plt.xlabel("Cities")
plt.ylabel("Number of Bars")
```
